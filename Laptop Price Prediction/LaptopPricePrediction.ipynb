```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df = df.iloc[:,2:]
```

```python
df.shape
```

```python
df.sample(5)
```

```python
df['OS'] = df['OS'].replace('Windows 11  OS','Windows 11 OS')
df['OS'] = df['OS'].replace('Windows 10  OS','Windows 10 OS')
df['OS'] = df['OS'].replace('Mac 10.15.3\t OS','Mac 10.15.3 OS')
df['Ram_type'] = df['Ram_type'].replace('LPDDR4x','LPDDR4X')
df['Ram_type'] = df['Ram_type'].replace('LPDDR5x','LPDDR5X')
df['processor']=df['processor'].str.split(' ').str.slice(0,5).str.join(' ')
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop('name',axis=1,inplace=True)
df.drop('GPU',axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop('price',axis=1),
                                                df['price'],
                                                test_size=0.2,
                                                random_state=13)
```

```python
brand_count=df['brand'].value_counts()
thresold = 15

repl = brand_count[brand_count <thresold].index
repl
```

```python
os_count= df['OS'].value_counts()

os_repl = os_count[os_count<10].index
os_repl
os_count
```

```python
pro_count=df['processor'].value_counts()
pro_repl=pro_count[pro_count<10].index
pro_repl
```

```python
cpu_count = df['CPU'].value_counts()
cpu_repl = cpu_count[cpu_count<10].index
cpu_repl
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
```

```python
def replace_band(column):
    col = pd.DataFrame(column)
    return col.replace(repl,'unknown')
```

```python
def replace_os(column):
    col = pd.DataFrame(column)
    return col.replace(os_repl,'unknown')
```

```python
def replace_cpu(column):
    col = pd.DataFrame(column)
    return col.replace(cpu_repl,'unknown')
```

```python
def replace_pro(column):
    col = pd.DataFrame(column)
    return col.replace(pro_repl,'unknown')
```

```python
t1 = ColumnTransformer([
    ('replace_brand',FunctionTransformer(func=replace_band,validate=False),[0]),
    ('replace_pro',FunctionTransformer(func=replace_pro,validate=False),[2]),
    ('replace_cpu',FunctionTransformer(func=replace_cpu,validate=False),[3]),
    ('replace_os',FunctionTransformer(func=replace_os,validate=False),[11])
],remainder='passthrough')
```

```python
t2 = ColumnTransformer([
    ('ohe',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[0,1,3])
],remainder='passthrough')

t3 = ColumnTransformer([
    ('oe_cpu',OrdinalEncoder(categories=[['10 Cores (6P + 4E), 16 Threads','Dual Core, 2 Threads','Dual Core, 4 Threads','Quad Core, 8 Threads','Hexa Core (2P + 4E), 8 Threads','Hexa Core, 12 Threads',
                                         'Octa Core (4P + 4E), 12 Threads','14 Cores (6P + 8E), 20 Threads','16 Cores (8P + 8E), 24 Threads','10 Cores (2P + 8E), 12 Threads','12 Cores (4P + 8E), 16 Threads','Octa Core, 16 Threads','24 Cores (8P + 16E), 32 Threads','unknown']]),[33]),
    ('oe_ram',OrdinalEncoder(categories=[['2GB','4GB','8GB','12GB','16GB','32GB','64GB']]),[35]),
    ('oe_ram_type',OrdinalEncoder(categories=[['DDR','DDR3','Unified','LPDDR4','DDR4-','LPDDR4X','LPDDR5','DDR4','LPDDR5X','DDR5']]),[36]),
    ('oe_rom',OrdinalEncoder(categories=[['32GB','64GB','128GB','256GB','512GB','1TB','2TB']]),[37]),
    ('oe_rom_type',OrdinalEncoder(categories=[['Hard-Disk','SSD']]),[38]),
    ('oe_warranty',OrdinalEncoder(categories=[[0,3,2,1]]),[42])
    
],remainder='passthrough')

t4 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,42))
],remainder='passthrough')

t5 = LinearRegression()

pipe = Pipeline([
    ('t1',t1),
    ('t2',t2),
    ('t3',t3),
    ('t4',t4),
    ('t5',t5)
])

pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
